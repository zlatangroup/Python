```python
car_sales=pd.read_csv("7.1 car-sales.csv")
```

```python
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].str.replace('[\$\,\.]','')
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].str[:-2]
```

```python
car_sales["Sales Date"]=pd.date_range('1/1/2020',periods=len(car_sales))
car_sales
```

```python

car_sales["Price"]=car_sales["Price"].astype(int)
car_sales["Total Sales"]=car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
car_sales.plot(x='Sales Date',y='Total Sales');
```

```python
car_sales.plot(x='Odometer (KM)',y='Price', kind="scatter");
```

```python
heart_disease=pd.read_csv("11.3 heart-disease.csv")
```

```python
heart_disease
```

```python
heart_disease["age"].plot.hist(
                            );
```

```python
heart_disease.plot.hist(figsize=(10,30),subplots=True);
```

```python
over_50=heart_disease[heart_disease["age"]>50]
```

```python
over_50.head()
```

```python
over_50.plot(kind="scatter",x="age",y="chol",c="target",figsize=(10,5))
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",x="age",y="chol",c="target",figsize=(10,5),ax=ax);
ax.set_xlim([45,100]);
```

```python
##using OOp 
fig,ax=plt.subplots(figsize=(10,6))
##plot data
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"]);
ax.set(title="Heart Disease and Cholersterol level",
       xlabel="Age",
       ylabel="Cholesterol");
ax.legend(*scatter.legend_elements(),title="target");
ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
over_50.head()
```

```python
fig,(ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(10,10))
scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])
#Customize axo
ax0.set(title="Heart Disease and Cholesterl",xlabel="Age",ylabel="Cholesterol");
#Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title='Target');
ax0.axhline(y=over_50["chol"].mean(),linestyle="--");

##add data to ax1
scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])
##Add title
ax1.set(title="Heart Disease and Max Heart Rate",xlabel="Age",ylabel="Heart rate")

#Add Legend to ax1

ax1.legend(*scatter.legend_elements(),title="Target");
ax1.axhline(y=over_50["thalach"].mean(),linestyle="--");
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");


```

```python
plt.style.use('seaborn-whitegrid')
##using OOp 
fig,ax=plt.subplots(figsize=(10,6))
##plot data
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"],
                  cmap="winter");
ax.set(title="Heart Disease and Cholersterol level",
       xlabel="Age",
       ylabel="Cholesterol");
ax.set_xlim([50,80])
ax.legend(*scatter.legend_elements(),title="target");
ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
fig,(ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(10,10))
scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"],
                   cmap="winter")
#Customize axo
ax0.set(title="Heart Disease and Cholesterol",xlabel="Age",ylabel="Cholesterol");
ax0.set_xlim([50,80])
#Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title='Target');
ax0.axhline(y=over_50["chol"].mean(),linestyle="--");

##add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"],
                    cmap="winter")
##Add title
ax1.set(title="Heart Disease and Max Heart Rate",xlabel="Age",ylabel="Heart rate")
#Change x axis limit
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
#Add Legend to ax1

ax1.legend(*scatter.legend_elements(),title="Target");
ax1.axhline(y=over_50["thalach"].mean(),linestyle="--");
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");

```
